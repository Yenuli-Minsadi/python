```python
school_clubs = {
    "Robotics" : ["Alice","Bob","Charlie","David","Eve"],
    "Chess" : ["Charlie","Eve","Frank","Grace"]
}
```

```python
# del set
robo_set = set(school_clubs["Robotics"])
chess_set = set(school_clubs["Chess"])

print(robo_set)
print(chess_set)
```

```python
only_robo = robo_set - chess_set
print(only_robo)
```

```python
all_students = robo_set | chess_set
print(all_students)
```
