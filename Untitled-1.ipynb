```python
x = 5

match x:
    case 1:
        print("x is 1")
    case 2:
        print("x is 2")
    case 3: 
        print("x is 3")
    case _:
        print("x is something else")
```

```python
my_list = [2, 4, 3]

match my_list:
    case [x, y]:
        print("x is", x)
        print("y is", y)
    case [x, y, z]:
        print("x is", x)
        print("y is", y)
        print("z is", z)
    case _:
        print("List has a different length")
```

```python

list = ["circle",4]

match list:
    case [c,r]:
        circle_area = 2 * 22 / 7 * r
        print(circle_area)
    case [s,l]:
        square_area = l ** 2
        print(square_area)
    case [r,l,b]:
        rectangle_area = l * b
        print(rectangle_area)
    case _:
        print("unknown shape or dimension")
        
```
